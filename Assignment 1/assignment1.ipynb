### Part 2: Preparation and transformation of the datasets
_This part of the notebook is doing simple data transformation on the training and testing dataset to prepare it for the model._

```python
# this cell is performing the scaling of the features
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Create an instance of MinMaxScaler and StandardScaler
min_max_scaler = MinMaxScaler()
standard_scaler = StandardScaler()

# create a dataframe simply for the 9 features
df_training_features = df_training_cleaned.filter(regex='^b', axis=1)
df_testing_features = df_testing_cleaned.filter(regex='^b', axis=1)
training_labels = df_training_cleaned['class']
testing_labels = df_testing_cleaned['class']

#fit the MinMaxScaler and StandardScaler based on the traing set
min_max_scaler.fit(df_training_features)
standard_scaler.fit(df_training_features)

#scale the features using the fit on training set

#scale the two datasets using the fitted MinMaxScaler, the results are returned in Numpy arrays
arr_training_features_min_max_scaled = min_max_scaler.transform(df_training_features)
arr_testing_features_min_max_scaled = min_max_scaler.transform(df_testing_features)

#scale the two datasets using the fitted StandardScaler, the results are returned in Numpy arrays
arr_training_features_standard_scaled = standard_scaler.transform(df_training_features)
arr_testing_features_standard_scaled = standard_scaler.transform(df_testing_features)

#converting the scaled features back into the dataframe

#for the minmax scaler, generate training and testing dataframes
df_training_min_max = pd.concat([pd.DataFrame(arr_training_features_min_max_scaled, columns=df_training_features.columns), training_labels.reset_index(drop=True)], axis=1)
df_testing_min_max = pd.concat([pd.DataFrame(arr_testing_features_min_max_scaled, columns=df_testing_features.columns), testing_labels.reset_index(drop=True)], axis=1)

#for the standard scaler, generate training and testing dataframes
df_training_standard = pd.concat([pd.DataFrame(arr_training_features_standard_scaled, columns=df_training_features.columns), training_labels.reset_index(drop=True)], axis=1)
df_testing_standard = pd.concat([pd.DataFrame(arr_testing_features_standard_scaled, columns=df_testing_features.columns), testing_labels.reset_index(drop=True)], axis=1)
```

To prevent data leakage, the scaler is only fit on the training set. The training and testing set is then transformed using the fitted scaler.

```python
# filter the training and testing dataframe to only have labels 's' and 'd'
df_training_min_max_binary = df_training_min_max.loc[df_training_min_max['class'].isin(['s','d'])]
df_testing_min_max_binary = df_testing_min_max.loc[df_testing_min_max['class'].isin(['s','d'])]

#avoid slicing on original copy error
df_training_min_max_binary = df_training_min_max_binary.copy()
df_testing_min_max_binary = df_testing_min_max_binary.copy()

# set class 'd' to be 0, class 's' to be 1
df_training_min_max_binary['class'] = df_training_min_max_binary['class'].apply(lambda x: 1 if x == 'd' else 0)
df_testing_min_max_binary['class'] = df_testing_min_max_binary['class'].apply(lambda x: 1 if x == 'd' else 0)
```

#### First, train the model for two individual features.

Before training the model on any features, three simple methods are used to check feature importance:
* Feature importance from a decision tree-based model: Random Forest or Gradient Boosting Trees
* Feature importance from a linear model: Logistic Regression with LASSO regularization)
* Permutation Importance

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.inspection import permutation_importance
from sklearn.metrics import accuracy_score, classification_report
import numpy as np

# The model is trained using the Min Max scaling dataset

# Split the training and testing dataset into features (X) and target labels (y)
df_training_min_max_binary_x = df_training_min_max_binary.loc[ : , ~df_training_min_max_binary.columns.str.startswith('c')]
df_training_min_max_binary_y = df_training_min_max_binary.loc[ : , df_training_min_max_binary.columns.str.startswith('c')]

df_testing_min_max_binary_x = df_testing_min_max_binary.loc[ : , ~df_testing_min_max_binary.columns.str.startswith('c')]
df_testing_min_max_binary_y = df_testing_min_max_binary.loc[ : , df_testing_min_max_binary.columns.str.startswith('c')]

# 1. Feature importance from a decision tree-based model (Random Forest)
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(df_training_min_max_binary_x, np.ravel(df_training_min_max_binary_y))
rf_feature_importances = rf.feature_importances_

# 2. Feature importance from a linear model (Logistic Regression with LASSO regularization)
log_reg = LogisticRegression(penalty='l1', solver='liblinear', random_state=42)
log_reg.fit(df_training_min_max_binary_x, np.ravel(df_training_min_max_binary_y))
log_reg_feature_importances = np.abs(log_reg.coef_).flatten()


# 3. Permutation Importance
permutation_imp = permutation_importance(log_reg, df_training_min_max_binary_x, np.ravel(df_training_min_max_binary_y), n_repeats=30, random_state=42)
perm_feature_importances = permutation_imp.importances_mean


# Print the feature importances
print("Random Forest feature importances:\n", rf_feature_importances)
print("Logistic Regression with LASSO feature importances:\n", log_reg_feature_importances)
print("Permutation Importance feature importances:\n", perm_feature_importances)

features = df_training_min_max_binary_x.columns

plt.subplot(3, 1, 1)
plt.plot(features,rf_feature_importances, 'r-')
plt.subplot(3, 1, 2)
plt.plot(features,log_reg_feature_importances, 'b-')
plt.subplot(3, 1, 3)
plt.plot(features,perm_feature_importances, 'g-')
plt.show()
```

```python
def plot_decision_boundary(arr_x_new,arr_y_proba,int_boundary,arr_testing_y,arr_testing_x,arr_test_proba,actual_or_prediction):
    #plot probability curve
    plt.plot(arr_x_new[:], arr_y_proba[:, 0], "g-", label="d")
    plt.plot(arr_x_new[:], arr_y_proba[:, 1], "b--", label="s")

    #plot vertical line for decision boundary
    plt.axvline(x = int_boundary, 
                color = 'black', 
                label = 'Decision Boundary',
                linestyle='dashdot')
        
    # plot the testing set on the graph, using their feature value and predicted probability to be 'd'
    plt.scatter(arr_testing_x[np.where(arr_testing_y==0)[0]], 
                arr_test_proba[:,0][np.where(arr_testing_y==0)[0]], 
                color='green', 
                label='Testing_'+ actual_or_prediction + '_d',
                s=12)
    plt.scatter(arr_testing_x[np.where(arr_testing_y==1)[0]], 
                arr_test_proba[:,0][np.where(arr_testing_y==1)[0]], 
                color='blue', 
                label='Testing_'+ actual_or_prediction + '_s',
                s=12)

    plt.legend(loc = 'upper right')
```

```python
def plot_proba_vs_score(scores_new,arr_y_proba,int_boundary,scores_test,arr_test_proba,arr_testing_y,actual_or_pred):
    # Plot the estimated probabilities vs. scores
    plt.plot(scores_new,
                arr_y_proba[:,0], 
                "g-", label="d")
    plt.plot(scores_new,
                arr_y_proba[:,1], 
                "b--", label="s")

    plt.axvline(x = int_boundary, 
                    color = 'black', 
                    label = 'Decision Boundary',
                    linestyle='dashdot')

    # plt.plot(arr_testing_x[:,i], arr_test_proba[:, 0], ".", label="Test_d_prob")
    plt.scatter(scores_test[np.where(arr_testing_y==0)[0]], 
                arr_test_proba[:,0][np.where(arr_testing_y==0)[0]], 
                color='green', 
                label='Testing_'+actual_or_pred+'_d',
                s=13)
    plt.scatter(scores_test[np.where(arr_testing_y==1)[0]], 
                arr_test_proba[:,0][np.where(arr_testing_y==1)[0]], 
                color='blue', 
                label='Testing_'+actual_or_pred+'_s',
                s=13)

    plt.xlabel('Scores (All features, linear part)')
    plt.ylabel('Estimated Probabilities')
    plt.legend()
    plt.title('Estimated Probabilities vs. Scores (All features, linear part)')
```

### Part 1: SVC classification
_This part of the code is implementing a Support Vector Machine Classifier with 3-fold cross validation._

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV


# Split the training and testing dataset into features (X) and target labels (y)
df_training_min_max_x = df_training_min_max.loc[ : , ~df_training_min_max.columns.str.startswith('c')]
df_training_min_max_y = df_training_min_max.loc[ : , df_training_min_max.columns.str.startswith('c')]

df_testing_min_max_x = df_testing_min_max.loc[ : , ~df_testing_min_max.columns.str.startswith('c')]
df_testing_min_max_y = df_testing_min_max.loc[ : , df_testing_min_max.columns.str.startswith('c')]

# Create an SVM classifier with one-versus-one strategy
svc_model  = SVC(decision_function_shape='ovo')

# Define the hyperparameters and their possible values for grid search
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf']
}

# Create a grid search object with 3-fold cross-validation
grid_search = GridSearchCV(svc_model, param_grid, cv=3, scoring='accuracy')

# Fit the grid search object to the training data
grid_search.fit(df_training_min_max_x, np.ravel(df_training_min_max_y))


# Print the best hyperparameters
print("Best hyperparameters:", grid_search.best_params_)

# Print the best cross-validation score
print("Best cross-validation score:", grid_search.best_score_)

# Evaluate the classifier with the best hyperparameters on the test set
best_svc = grid_search.best_estimator_
test_accuracy = best_svc.score(df_testing_min_max_x, np.ravel(df_testing_min_max_y))
print("Test set accuracy:", test_accuracy)
```

### Part 1: Softmax Regression and the k-NN algorithm
_This part of the code is implementing a Softmax Regression and the k-NN algorithm with 3-fold cross validation._

```python
softmax_reg = LogisticRegression(multi_class='multinomial', solver='lbfgs', random_state=42)
k_values = [1, 3, 5, 7, 9]
knn_classifiers = [KNeighborsClassifier(n_neighbors=k) for k in k_values]

# Fit the classifiers
softmax_reg.fit(df_training_min_max_x, np.ravel(df_training_min_max_y))
for knn in knn_classifiers:
    knn.fit(df_training_min_max_x, np.ravel(df_training_min_max_y))

# Evaluate the classifiers using cross-validation
softmax_reg_scores = cross_val_score(softmax_reg, df_training_min_max_x, np.ravel(df_training_min_max_y), cv=3)
knn_scores = [cross_val_score(knn, df_training_min_max_x, np.ravel(df_training_min_max_y), cv=3) for knn in knn_classifiers]


# Print the average cross-validation scores
print("Softmax Regression average CV score:", softmax_reg_scores.mean())
for k, scores in zip(k_values, knn_scores):
    print(f"k-NN (k={k}) average CV score:", scores.mean())
```

# Section 3: California Housing Data

### Part 1: Load housing Data
_Housing data is loaded using the codes provided from lecture 2._

```python
import os
import tarfile
import urllib.request

DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml2/master/"
HOUSING_PATH = os.path.join("datasets", "housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()

fetch_housing_data()
```

```python
import pandas as pd

def load_housing_data(housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path, "housing.csv")
    return pd.read_csv(csv_path)

df_housing = load_housing_data()
df_housing.head()
```
