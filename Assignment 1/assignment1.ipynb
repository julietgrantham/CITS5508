### Part 2: Preparation and transformation of the datasets
_This part of the notebook is doing simple data transformation on the training and testing dataset to prepare it for the model._

```python
# this cell is performing the scaling of the features
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Create an instance of MinMaxScaler and StandardScaler
min_max_scaler = MinMaxScaler()
standard_scaler = StandardScaler()

# create a dataframe simply for the 9 features, removing the target class
df_training_features = df_training_cleaned.filter(regex='^b', axis=1)
df_testing_features = df_testing_cleaned.filter(regex='^b', axis=1)
training_labels = df_training_cleaned['class']
testing_labels = df_testing_cleaned['class']

#fit the MinMaxScaler and StandardScaler based on the traing set
min_max_scaler.fit(df_training_features)
standard_scaler.fit(df_training_features)

#scale the features using the fit on training set

#scale the two datasets using the fitted MinMaxScaler
arr_training_features_min_max_scaled = min_max_scaler.transform(df_training_features)
arr_testing_features_min_max_scaled = min_max_scaler.transform(df_testing_features)

#scale the two datasets using the fitted StandardScaler
arr_training_features_standard_scaled = standard_scaler.transform(df_training_features)
arr_testing_features_standard_scaled = standard_scaler.transform(df_testing_features)

#converting the scaled features back into the dataframe

#for the minmax scaler, generate training and testing dataframes
df_training_min_max = pd.concat([pd.DataFrame(arr_training_features_min_max_scaled, columns=df_training_features.columns), training_labels.reset_index(drop=True)], axis=1)
df_testing_min_max = pd.concat([pd.DataFrame(arr_testing_features_min_max_scaled, columns=df_testing_features.columns), testing_labels.reset_index(drop=True)], axis=1)

#for the standard scaler, generate training and testing dataframes
df_training_standard = pd.concat([pd.DataFrame(arr_training_features_standard_scaled, columns=df_training_features.columns), training_labels.reset_index(drop=True)], axis=1)
df_testing_standard = pd.concat([pd.DataFrame(arr_testing_features_standard_scaled, columns=df_testing_features.columns), testing_labels.reset_index(drop=True)], axis=1)
```

##### Simple feature importance calculations for pre-checks.
Before training the model on any features, three simple methods are used to check feature importance:
* Feature importance from a decision tree-based model: Random Forest or Gradient Boosting Trees
* Feature importance from a linear model: Logistic Regression with LASSO regularization
* Permutation Importance

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.inspection import permutation_importance
from sklearn.metrics import accuracy_score, classification_report
import numpy as np

# The model is trained using the Min Max scaling dataset

# Split the training and testing dataset into features (X) and target labels (y)
df_training_min_max_binary_x = df_training_min_max_binary.loc[ : , ~df_training_min_max_binary.columns.str.startswith('c')]
df_training_min_max_binary_y = df_training_min_max_binary.loc[ : , df_training_min_max_binary.columns.str.startswith('c')]

df_testing_min_max_binary_x = df_testing_min_max_binary.loc[ : , ~df_testing_min_max_binary.columns.str.startswith('c')]
df_testing_min_max_binary_y = df_testing_min_max_binary.loc[ : , df_testing_min_max_binary.columns.str.startswith('c')]

# Method 1: Feature importance from a decision tree-based model (Random Forest)
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(df_training_min_max_binary_x, np.ravel(df_training_min_max_binary_y))
rf_feature_importances = rf.feature_importances_

# Method 2: Feature importance from a linear model (Logistic Regression with LASSO regularization)
log_reg = LogisticRegression(penalty='l1', solver='liblinear', random_state=42)
log_reg.fit(df_training_min_max_binary_x, np.ravel(df_training_min_max_binary_y))
log_reg_feature_importances = np.abs(log_reg.coef_).flatten()


# Method 3: Permutation Importance
permutation_imp = permutation_importance(log_reg, df_training_min_max_binary_x, np.ravel(df_training_min_max_binary_y), n_repeats=30, random_state=42)
perm_feature_importances = permutation_imp.importances_mean


# Print the feature importances
print("Random Forest feature importances:\n", rf_feature_importances)
print("Logistic Regression with LASSO feature importances:\n", log_reg_feature_importances)
print("Permutation Importance feature importances:\n", perm_feature_importances)

features = df_training_min_max_binary_x.columns

plt.subplot(3, 1, 1)
plt.plot(features,rf_feature_importances, 'r-')
plt.subplot(3, 1, 2)
plt.plot(features,log_reg_feature_importances, 'b-')
plt.subplot(3, 1, 3)
plt.plot(features,perm_feature_importances, 'g-')
plt.show()
```

Note that, there are many parameters of logistic regression that can be defined here:
* penalty: Default is 'l2' (Ridge). It can also be 'l1' (lasso), none, or 'elasticnet'
* dual: Default is False. Dual or primal formulation. It is better to set dual = false (default) where there are more samples than features. The dual formulation is also only available to be used in combination with 'l2' and 'liblinear' solver
* tolL Tolerance for the stopping criteria. It is default to be 0.0001
* C: It defines the regularization strength. It is default to be 1. The higher the value, the more regularization and the more rigid decision boundary.  
* fit_intercept: Defined if there can be a bias term. It is default to be True.
* intercept_scaling: Default to be 1. It can only be used with 'liblinear' solver for synthetic features. It needs to be connsidered together with the regularization terms. 
* class_weight: Default is none, assuming all calss have weight 1.
* random_state: Defaultis none. Some Solvers shuffles the data. 
* solver: Default to be'lbfgs'. Can have values: 'lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'.
* max_iter: Defaults to be 100. Maximum iterations for the solver to converge.
* multi_class: Defaults to be 'auto'. Can have values: 'auto', 'ovr', 'multinomial'. 
* verbose: Defaults to be 0.
* warm_start: Defaults to be False. Can be set to true to reuse the previous solutions for initial fitting.
* n_jobs: Defaults to be None. Specified number of cpu cores if multi_class is 'ovr'. 
* l1_ratio: Defaults to be None.  Only used when penalty is 'elasticnet'

##### Plots and discussions of probability curves and decision boundaries

```python
# Defines the function that plots probability curves and the decision boundary
def plot_decision_boundary(arr_x_new,arr_y_proba,int_boundary,arr_testing_y,arr_testing_x,arr_test_proba,actual_or_prediction):
    #plot probability curve
    plt.plot(arr_x_new[:], arr_y_proba[:, 0], "g-", label="Prob_d")
    plt.plot(arr_x_new[:], arr_y_proba[:, 1], "b--", label="Prob_s")

    #plot vertical line for decision boundary
    plt.axvline(x = int_boundary, 
                color = 'black', 
                label = 'Decision Boundary',
                linestyle='dashdot')
        
    # plot the testing set on the graph, 
        #if the label (actual or prediction) is 'd', using the probability of 'd'
        #if the label (actual or prediction) is 's', using the probability of 's'
    plt.scatter(arr_testing_x[np.where(arr_testing_y==0)[0]], 
                arr_test_proba[:, 0][np.where(arr_testing_y==0)[0]], 
                color='green', 
                label='Testing_'+ actual_or_prediction + '_d',
                s=5)
    plt.scatter(arr_testing_x[np.where(arr_testing_y==1)[0]], 
                arr_test_proba[:,1][np.where(arr_testing_y==1)[0]], 
                color='blue', 
                label='Testing_'+ actual_or_prediction + '_s',
                s=5)

#show legend upper right corner
    plt.legend(loc = 'center right')

    plt.ylabel('Probability')
    plt.xlabel('Feature Value')
```

```python
# Define function to plot Precision, Recall vs, threshold
def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):
    # Get the recall value where precision is above 0.9
    recall_90_precision = recalls[np.argmax(precisions >= 0.90)]
    # Get threshold valie where precision is above 0.9
    threshold_90_precision = thresholds[np.argmax(precisions >= 0.90)]


    plt.plot(thresholds, precisions[:-1], "b--", label="Precision", linewidth=2)
    plt.plot(thresholds, recalls[:-1], "g-", label="Recall", linewidth=2)
    plt.legend(loc="center right", fontsize=10) 
    plt.xlabel("Threshold", fontsize=10)       
    plt.grid(True)                              
    plt.axis([0, 1, 0, 1.1])             
    #plot vertical dash line from precision curve where precision is 0.9
    plt.plot([threshold_90_precision, threshold_90_precision], [0., 0.9], "r:")  

    #plot horizotal dash line from precision curve where precision is 0.9
    plt.plot([-1, threshold_90_precision], [0.9, 0.9], "r:")     

    #plot horisontal line from recall curve where precision is 0.9
    plt.plot([-1, threshold_90_precision], [recall_90_precision, recall_90_precision], "r:")

    #plot dot on precision curve where precision is 0.9
    plt.plot([threshold_90_precision], [0.9], "ro")         

    #plot dot on recall curve where precision is 0.9
    plt.plot([threshold_90_precision], [recall_90_precision], "ro")   
```

In reference to the graphs above:
* The four graphs above is the precision & recall VS. threshold for the four models trained: Three on individual feature b1, b2, b5, and one on all features. 
* Precision refers to the ratio of true positives within all the indentified positives (TP/(TP + FP))
* Recall refers to the ratio of all the actual positives, how many are identified as true positive (TP/(TP + FN))
* Usually there is a trade-off between high precision and high recall. The two cannot be achieved at the same time. 
* As can be seen from the graphs above, the trade-off is visible in all four models to different extend. Model trained on b1 has the worst trade-off, either one drops significantly as the other one increases. Model trained in all features has the best performance, the trade-off drops the slowest as the threshold increases and decreases. 
* The performance of percision recall trade off is inate to the model.

To find the best threshold:
* It depends on the application of the model -- to have high precision (less false positives) or to have high recall (less false negatives). 
* For example, in medical context of screening for disease. For severe diseases with high fatality, we would like to find all people that possibly has the disease and not miss out false negatives. In this case, a low threshold for high recall is needed. For less severe diseases that is common in population, a higher precision is needed to have less false positive cases to reduce the workload of the hispital. In this case, a higher threshold is recommended to have higher precision. 

In term of this question, below, we are using the method to find the threshold that produces the best balance between precision and recall. It is discussed in more details in the later on.

The below is finding the threshold that will lead to the smallest difference between precision and recall for each one of the four models.

```python
# Find the threshold that balances precision and recall

# for b1 model
optimal_idx_b1 = np.argmin(np.abs(precisions_log_b1 - recalls_log_b1))
optimal_threshold_b1 = thresholds_log_b1[optimal_idx_b1]

# for b2 model
optimal_idx_b2 = np.argmin(np.abs(precisions_log_b2 - recalls_log_b2))
optimal_threshold_b2 = thresholds_log_b2[optimal_idx_b2]

# for b5 model
optimal_idx_b5 = np.argmin(np.abs(precisions_log_b5 - recalls_log_b5))
optimal_threshold_b5 = thresholds_log_b5[optimal_idx_b5]

# for all features model
optimal_idx = np.argmin(np.abs(precisions_log - recalls_log))
optimal_threshold = thresholds_log[optimal_idx]
```

```python
# Define function to plot Precision VS. Recall 
def plot_precision_vs_recall(precisions, recalls, optimal_idx,thresholds):
    plt.plot(recalls, precisions, linewidth=2)
    plt.xlabel("Recall", fontsize=10)
    plt.ylabel("Precision", fontsize=10)
    plt.axis([0, 1.05, 0, 1.05])
    # Get the recall value where precision is above 0.9
    recall_90_precision = recalls[np.argmax(precisions >= 0.90)]

    #plot where 90% precision is
    plt.plot([recall_90_precision, recall_90_precision], [0., 0.9], "r:")
    plt.plot([0.0, recall_90_precision], [0.9, 0.9], "r:")
    plt.plot([recall_90_precision], [0.9], "ro", label = '90% Precision Threshold: ' + str(round(thresholds[np.argmax(precisions >= 0.90)],2)))

    # plot most balanced precision & recall
    plt.plot([0.0, recalls[optimal_idx]], [precisions[optimal_idx], precisions[optimal_idx]], "g:")
    plt.plot([recalls[optimal_idx],recalls[optimal_idx]], [precisions[optimal_idx],0.0], "g:")
    plt.plot(recalls[optimal_idx], precisions[optimal_idx], "go", label = 'Most Balanced Threshold: ' + str(round(thresholds[optimal_idx],2)))

    plt.legend(loc='lower left')
    plt.grid(True)
```

```python
def plot_against_num_neighbor(k_values,li_mean_knn_scores,softmax_reg_scores,grid_search_simple_score,grid_search_score,acc_val):
    plt.plot(k_values,li_mean_knn_scores,'b--', label ='KNN '+ acc_val + ' Curve')
    # mark score for softmax
    plt.axhline(y = softmax_reg_scores.mean(), 
                color = 'green', 
                label = 'Softmax ' + acc_val +': ' + str(round(softmax_reg_scores.mean(),2)),
                linestyle='dashdot')

    # mark score for linear SVM
    plt.axhline(y = grid_search_simple_score, 
                color = 'red', 
                label = 'SVM Linear ' + acc_val +': ' + str(round(grid_search_simple_score,2)),
                linestyle='dashdot')

    # mark score for polynomial svm
    plt.axhline(y = grid_search_score, 
                color = 'orange', 
                label = 'SVM Poly ' + acc_val +': ' + str(round(grid_search_score,2)),
                linestyle='dashdot')

    # mark maximum knn score
    plt.axhline(y = max(li_mean_knn_scores), 
                color = 'blue', 
                label = 'KNN Max ' + acc_val +': ' + str(round(max(li_mean_knn_scores),2)),
                linestyle='dashdot')
    plt.legend()
    plt.title(acc_val + ' comparison')
    plt.xlabel('Number of neighbors')
    plt.ylabel(acc_val)
```

```python
# Train k-NN regressors with 'uniform' and 'distance' weights and 9 neighbors
knn_uniform = KNeighborsRegressor(n_neighbors=9, weights='uniform')
knn_distance = KNeighborsRegressor(n_neighbors=9, weights='distance')

# fit the datas
knn_uniform.fit(df_X_train_min_max_scaled.loc[:,['median_income','ocean_proximity','population', 'longitude','latitude']], y_train)
knn_distance.fit(df_X_train_min_max_scaled.loc[:,['median_income','ocean_proximity','population', 'longitude','latitude']], y_train)

# Make predictions on the test set
y_pred_uniform = knn_uniform.predict(df_X_test_min_max_scaled.loc[:,['median_income','ocean_proximity','population', 'longitude','latitude']])
y_pred_distance = knn_distance.predict(df_X_test_min_max_scaled.loc[:,['median_income','ocean_proximity','population', 'longitude','latitude']])

# Compute the RMSE
rmse_uniform = np.sqrt(mean_squared_error(y_test, y_pred_uniform))
print("RMSE with uniform weights is:", rmse_uniform)

rmse_distance = np.sqrt(mean_squared_error(y_test, y_pred_distance))
print("RMSE with distance weights is:", rmse_distance)
```

AS can be seen from above:
* The default weights used it 'uniform'
* Having 'distance' weights improves the model performance, as it takes into account the distance of the neighbors to give them different weights.<br>

Custom weights functions can also be used here for custom weight definition. Some examples shown below.

```python
# Linear decay weights
def linear_decay(distances):
    max_distance = np.max(distances)
    weights = 1 - (distances / max_distance)
    return weights

# This is the same as the 'distance' weights
def inverse_distance(distances):
    weights = 1 / distances
    return weights

# Gaussian weights, width = 1
def gaussian_kernel(distances, sigma=1):
    weights = np.exp(-distances**2 / (2 * sigma**2))
    return weights
```

```python
# Train k-NN regressors with 'linear_decay', 'inverse_distance' (same as 'distance') and 'gaussian_kernel' weights
knn_linear_decay = KNeighborsRegressor(n_neighbors=9, weights=linear_decay)
knn_inverse_distance = KNeighborsRegressor(n_neighbors=9, weights=inverse_distance)
knn_gaussian_kernel = KNeighborsRegressor(n_neighbors=9, weights=gaussian_kernel)

# fit the training data
knn_linear_decay.fit(df_X_train_min_max_scaled.loc[:,['median_income','ocean_proximity','population', 'longitude','latitude']], y_train)
knn_inverse_distance.fit(df_X_train_min_max_scaled.loc[:,['median_income','ocean_proximity','population', 'longitude','latitude']], y_train)
knn_gaussian_kernel.fit(df_X_train_min_max_scaled.loc[:,['median_income','ocean_proximity','population', 'longitude','latitude']], y_train)

# Make predictions on the test set
y_pred_linear_decay = knn_linear_decay.predict(df_X_test_min_max_scaled.loc[:,['median_income','ocean_proximity','population', 'longitude','latitude']])
y_pred_inverse_distance = knn_inverse_distance.predict(df_X_test_min_max_scaled.loc[:,['median_income','ocean_proximity','population', 'longitude','latitude']])
y_pred_gaussian_kernel = knn_gaussian_kernel.predict(df_X_test_min_max_scaled.loc[:,['median_income','ocean_proximity','population', 'longitude','latitude']])

# Compute the RMSE
rmse_linear_decay = np.sqrt(mean_squared_error(y_test, y_pred_linear_decay))
print("RMSE with linear_decay weights is:", rmse_linear_decay)

# Compute the RMSE
rmse_inverse_distance = np.sqrt(mean_squared_error(y_test, y_pred_inverse_distance))
print("RMSE with inverse_distance weights is:", rmse_inverse_distance)

# Compute the RMSE
rmse_gaussian_kernel = np.sqrt(mean_squared_error(y_test, y_pred_gaussian_kernel))
print("RMSE with gaussian_kernel weights is:", rmse_gaussian_kernel)
```

As can be seen from above:
* The inverse_distance weights is the definition of 'distance' weights
* The inverse_distance still have the best performance. 
* Although linear_decay and gaussian_kernel does not have as good performance as inverse_distance, they both perform better than 'uniform' weights. (not by much)
